This is just a short notebook to play with logs and viewing files that are produced. You can also see the logs under:

```
Clusters -> Driver Logs
```

but I have found them easier to track and view through this notebook.

````
# Set variables
clustername='jr-rstudio-tst2'# name of cluster
dir_idx_to_look_at = -1     # directory to look in for init log files, default (-1) should be the **last/most recent**
file2lookatidx = 0          # file within that directory to look at.
````

````
## import os to join dirs appropriately
import os
## get all logdirs from init scripts, and print out the last one:
clusterpath = os.path.join('/databricks/init/output',clustername)
logdirs = dbutils.fs.ls(clusterpath)
display(logdirs)# display all
````

````
print("""
********
Processing Directory: %s
********
"""%logdirs[dir_idx_to_look_at].path)
````

````
files=dbutils.fs.ls(logdirs[dir_idx_to_look_at].path)
## There should be N where N is the number of machines in the cluster.
display(files)
````

````
print(files[file2lookatidx].path)
````
